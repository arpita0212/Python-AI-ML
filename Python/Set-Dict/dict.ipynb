```python
# <!-- list comprehensive -->

squares_even = {}
for i in range(1, 11):
    if i % 2 == 0:
        squares_even[i] = i ** 2

print(squares_even)
```

```python
#  without comprehension

squares_cubes = {}
for i in range(1, 11):
    if i % 2 == 0:
        squares_cubes[i] = i ** 2
    else: 
        squares_cubes[i] = i ** 3

print(squares_cubes)
```
