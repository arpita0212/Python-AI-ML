```python
# set is unordered, 

# creating a set
s = {131,232,21,"Arpita", "Raj", 42,42,12,42}
print(s)
print(type(s))

# print(s[0]) # set is unordered, indexing is not possible
# this will raise an error: TypeError: 'set' object is not subscriptable



# adding an element to a set
s.add(100)
print(s)

# adding multiple elements to a set
s.update([200,300,400]) # update() can take an iterable (like list, set, tuple) and add its elements to the set
print(s)

# updating a set with another set
s2 = {500,600}
s.update(s2)
print(s)

# removing an element from a set
s.remove(100) # remove() will raise an error if the element is not found
print(s)

s.discard(200) # discard() will not raise an error if the element is not found
print(s)
```
