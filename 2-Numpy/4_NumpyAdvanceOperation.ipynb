```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])


print("==== Zomato Sales analysis ====")
print("\n sales data shape", sales_data.shape)
print("\n sample data for 1st 3 restuants: \n", sales_data[:3])
print("\n sales data ", sales_data[:, 1:])
```

```python
# total sales per year

print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)

# axis=0 -> row
# axis=1 -> column
```

## Minimum sales per resturant

axis=0 → Down the rows (column-wise operation)
axis=1 → Across the columns (row-wise operation)

```python
min_sales = np.min(sales_data[:, 1:], axis=0)

```

### Maximum sales per year

```python
max_sales = np.max(sales_data[:, 1:], axis=0)
print("max sales ", max_sales)
```

Average sales per resturant

```python
avg_sales = np.mean(sales_data[:, 1:], axis=1)  #if we go left to right
print("avg sales", avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1) 
print("cumsum", cumsum)


plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title('Average Cumulative Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()


```
